```python
sns.displot(data=df_sanmartin,x='AVANCE_FISICO')
plt.show()
```

```python
torta = df_sanmartin.groupby(['SITUACION']).size().reset_index(name='counts')

pieChart = px.pie(torta, values='counts', names='SITUACION', 
                  title='Distribucion de tipo de produccion' ,
                  color_discrete_sequence=px.colors.qualitative.Set3)

pieChart.show()
```

```python
df_sanmartin['FUNCION']=df_sanmartin['FUNCION'].fillna(' No Especificado') # en el dataframe, enla columan country, llenamos los natos nulos por no especificado
datos_filtrados= pd.DataFrame()# convertimos nuestra nuevo data frame en uno nuevo
datos_filtrados= df_sanmartin['FUNCION'].str.split(',',expand=True).stack() #
datos_filtrados= datos_filtrados.to_frame()
datos_filtrados.columns=['Pais'] # Renombramos la columna

paises = datos_filtrados.groupby(['Pais']).size().reset_index(name='Total')
paises= paises[paises.Pais != 'Pais no Especificado']
paises= paises.sort_values(by=['Total'], ascending=False)

paisesTop5=paises.head()
paisesTop5=paisesTop5.sort_values(by=['Total'])

fig1= px.bar(paisesTop5, x='Pais', y='Total', title='Top 5 paises')
fig1.show()
```
